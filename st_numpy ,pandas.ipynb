```python
import numpy as np #넘파이는 행렬
```

```python
arr_1 = np.array(range(10)) #[1,2,3,4,5,6,7,8,9]
print(arr_1)
print(arr_1.shape)  #1,2,3..은 모두 행
print(f'arr_1의 차원: {arr_1.ndim}, \n\n') 

arr_2 = np.array([[1,2,3],[1,2,3]])
print(arr_2)
print(arr_2.shape)
print(f'arr_2의 차원: {arr_2.ndim}, \n\n')

arr_3 = np.array([[[1,2,3],[1,2,3]],[[1,2,3],[1,2,3]]])
print(arr_3)
print(arr_3.shape)
print(f'arr_3의 차원: {arr_3.ndim}, \n')
```

```python
arr1 = np.arange(start =2, stop = 20).reshape(-1,3) # -1을 넣으면 나머지 차원에 맞도록 알아서
print(arr1)
```

```python
print(arr1[2], '\n')
print(arr1[1:3],'\n') #행 슬라이싱
print(arr1[:,1:3]) #열 슬라이싱 하려면 @@@@  ',' 로 구분 후 행에 ':' 를 꼭 넣어줘야 함
```

```python
#위에처럼도 같음
arr1[3][2]
arr1[3,2]
```

```python
#불린 인덱싱 - 조건 검색
arr1d = arr1[arr1>10]
print(arr1d)
```

```python
# 정렬
arr1_desc = np.sort(arr1)[::-1]
print(arr1_desc)

arr1_asc = np.sort(arr1_desc)  #열 정렬이라 그대로 == axis=1
print(arr1_asc)

arr1_sort = np.sort(arr1_desc, axis=0) #행 정렬
print(arr1_sort)
```

```python
#결손 데이터 대체 / #실제 데이터를 변경하려면 fillna('C000', inplace=True)
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(4)
```

```python
#나이의 결측치에 나이의 평균(결측치 제외한) 
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
```

```python
#판다스식 lambda. ex) 'Name'의 글자길이로 새로운 컬럼 형성. ap을 사용하지 않음
titanic_df["Name_len"] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name', 'Name_len']].head()
```

```python
# X -> 각 행별 len이 아닌 {Name: [, , , , ]}의 값들 전체를 취급
# lambda를 사용해야 행별로 각각 계산해야 됨
titanic_df['Name_len'] = len(titanic_df['Name'])
titanic_df.head()
```

```python
#if를 사용
titanic_df['C/A'] = titanic_df['Age'].apply(lambda x: 'child' if x<=19 else 'adult')
titanic_df[10:15]
```

```python
#else if는 지원하지 않지만 if, else를 두 번 사용하면 가능
titanic_df['C/A/E'] = titanic_df['Age'].apply(lambda x: 'child' if x<=19 else ('adult' if x <= 61 else 'elderly'))
titanic_df.iloc[33:40]
```
